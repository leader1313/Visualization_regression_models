```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from Utils import ID_generator
ID = ID_generator()
```

```python
plt.style.use("ggplot")
X = np.linspace(0, 3*np.pi, num=100)
for i in range(5):
    Y = np.sin(X+ np.random.normal(loc=0.0, scale=1.0)) + np.random.normal(loc=0.0, scale=1.0)
    line = plt.plot(X,Y,linewidth=2)
    plt.fill_between(X,y1=Y + 0.5,y2=Y - 0.5, 
                        color=line[0].get_color(),
                        alpha=0.2)
yticks = np.linspace(-1,1,num=5)
plt.yticks(fontsize=0)
plt.xticks(fontsize=0)
plt.savefig("Figures/IOMGP_prior/{}.png".format(ID.now_stamp()))
```
