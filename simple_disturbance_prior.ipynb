```python
import numpy as np
import matplotlib.pyplot as plt
```

# state-independent disturbance prior

```python
plt.style.use("ggplot")
X = np.linspace(0, 3*np.pi, num=100)
Y = np.zeros(100)
Y1 = Y + 0.5
Y2 = Y - 0.5
plt.plot(X,Y,linewidth=2)
plt.fill_between(X,y1=Y1,y2=Y2, 
                    color="blue",
                    alpha=0.2,
                    hatch="\\",)
yticks = np.linspace(-1,1,num=5)
plt.yticks(yticks)
plt.savefig("Figures/Simple_disturbance_prior/state-independent_disturbance_prior.png")
```

# state-dependent disturbance prior

```python
plt.style.use("ggplot")
X = np.linspace(0, 3*np.pi, num=100)
Y = np.zeros(100)
Y1 = Y + np.sin(X)/4 + 0.5
Y2 = Y -  np.sin(X)/4 -0.5
plt.plot(X,Y,linewidth=2)
plt.fill_between(X,y1=Y1,y2=Y2, 
                    color="blue",
                    alpha=0.2,
                    hatch="\\",)
yticks = np.linspace(-1,1,num=5)
plt.yticks(yticks)
plt.savefig("Figures/Simple_disturbance_prior/state-dependent_disturbance_prior.png")
```
